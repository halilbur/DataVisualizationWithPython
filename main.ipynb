Take specific number of sample and construct %95 confidence
interval for the mean and variance. can you do it without 3rd party libraries and use my dataset

```python
sample_size = len(data_of_column)

critical_value = 1.96

margin_of_error = critical_value * std_error_of_column

lower_bound_mean = mean_of_column - margin_of_error
upper_bound_mean = mean_of_column + margin_of_error

lower_bound_variance = variance_of_column - margin_of_error
upper_bound_variance = variance_of_column + margin_of_error

print(f"Mean: {mean_of_column:.3f}")
print(f"Variance: {variance_of_column:.3f}")
print(f"95% Confidence Interval for the Mean: ({lower_bound_mean:.3f}, {upper_bound_mean:.3f})")
print(f"95% Confidence Interval for the Variance: ({lower_bound_variance:.3f}, {upper_bound_variance:.3f})")
```
